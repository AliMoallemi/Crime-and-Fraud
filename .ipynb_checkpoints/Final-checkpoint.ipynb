## Feature Engineering

### Date Time Data

```python
#change time stamp to multiple columns in order to make it more expressive for Later Classification
def DTConvertor(df, col):
    #imput is a data frame and a timestamp column
    #outputs the same dataframe with extra columns includin, day, hour, minute, weekday, day of week, etc.
    # For more Dertails Look at https://pandas.pydata.org/pandas-docs/stable/reference/series.html#api-series-dt
    df['Year'+"_"+col] = df[col].dt.year
    df['Month'+"_"+col] = df[col].dt.month
    df['Day'+"_"+col] = df[col].dt.day
    df['Hour'+"_"+col] = df[col].dt.hour
    df['Minute'+"_"+col] = df[col].dt.minute
    df['Week'+"_"+col] = df[col].dt.week
    df['WeekOfYear'+"_"+col] = df[col].dt.weekofyear
    df['DayOfWeek'+"_"+col] = df[col].dt.dayofweek
    df['Weekday'+"_"+col] = df[col].dt.weekday
    df['DayOfYear'+"_"+col] = df[col].dt.dayofyear
    df['Quarter'+"_"+col] = df[col].dt.quarter
    #df['is_month_start'+"_"+col] = df[col].dt.is_month_start
    #df['is_month_end'+"_"+col] = df[col].dt.is_month_end
    #df['is_quarter_start'+"_"+col] = df[col].dt.is_quarter_start
    #df['is_quarter_end'+"_"+col] = df[col].dt.is_quarter_end
    #df['is_year_start'+"_"+col] = df[col].dt.is_year_start
    #df['is_year_end'+"_"+col] = df[col].dt.is_year_end
    df['TimeZone'+"_"+col] = df[col].dt.tz
    df.drop(columns=col, inplace=True)
    return df
```

### Categorical Data

#### Categorical data are usually converted vai OneHotEncoder, LabelENcoder, etc But I prefer to use more statistical values which represent them better

```python
def CAtDescriber(df, col, n_top_cat=30):
    # Input: data frame, and dategorical Column and n_top_cat with default value 30
    #Output: counts the number of unique categories in the column and Plots it
    print(f'The number of unique categories in {col} is: ', df[col].nunique())
    main_categories= df[col].value_counts(normalize=True).reset_index()
    # we present the top n_top_cat categories 
    main_categories.set_index('index')[:n_top_cat].plot.pie(subplots=True, figsize = (10,10))
    plt.show()
```

```python
# a function to convert categorical data into more meaningful numerical ones
def CAtConvertor(df, big_df, cols, num_cols):
    # Input: data frame, and dategorical Column and n_top_cat with default value 30
    #Output: counts the number of unique categories in the column and Plots it
    agg_func = ['mean','std']
    
    for col in cols:
        desc_categories= big_df.groupby([col])[num_cols].agg(agg_func).fillna(0.0)
        desc_categories.columns = [col+"_"+'_and_'.join(colum).strip() for colum in desc_categories.columns.values]
        df = df.merge(desc_categories.drop_duplicates(), left_on=col, right_index=True, how = 'left').fillna(-1.0)
    return df
```

## Deep Neural Network Model

```python
from tensorflow import keras
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Activation, Input

from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

```

```python
er = EarlyStopping(monitor='val_loss', patience=10, verbose=1, mode='auto')
rl = ReduceLROnPlateau(monitor='val_loss', patience=3, verbose=1, mode='auto')
```

#### Neural Networks will help up to extract  nonlinear interdependencies

```python
def AUC_model(loss_b):
    model = Sequential()
    model.add(
        Dense(
            256,
            input_dim=X.shape[1],
            kernel_initializer='glorot_normal',
            ))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    model.add(Dense(256, kernel_initializer='glorot_normal'))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.25))
    model.add(Dense(64, kernel_initializer='glorot_normal'))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.25))
    model.add(Dense(64, kernel_initializer='glorot_normal'))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.25))
    model.add(Dense(16, kernel_initializer='glorot_normal'))
    model.add(Activation('relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.1))
    model.add(Dense(1, activation='sigmoid'))

    # Compile model
    model.compile(optimizer='Adam', metrics = ['AUC', 'accuracy'], loss=loss_b)
    print(model.summary())

    return model
```

```python
#Using KL-Divergence as Loss helps us to find the most probable False answers
Deep_Model = AUC_model('kullback_leibler_divergence')
```

```python
history_AUC_PCA = Deep_Model.fit(X_train_pca, y_train_pca, validation_data= (X_val_pca, y_val_pca), 
                                 epochs=2, batch_size=1000, callbacks=[er,rl],#class_weight=Counter(y_pca),
                                 verbose=2)
```

```python
def ROCPlotter(pred, y_true):
    #Input: prediction probabilities and true labels
    #Output: ROC AUC curve
    
    lr_probs = pred
    ns_probs = [0 for _ in range(len(y_true))]
    testy = y_true
    ns_auc = roc_auc_score(testy, ns_probs)
    lr_auc = roc_auc_score(testy, lr_probs)
    print('Logistic: ROC AUC=%.3f' % (lr_auc))
    # calculate roc curves
    ns_fpr, ns_tpr, _ = roc_curve(testy, ns_probs)
    lr_fpr, lr_tpr, _ = roc_curve(testy, lr_probs)
    # plot the roc curve for the model
    plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
    plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')
    # axis labels
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    # show the legend
    plt.legend()
    # show the plot
    plt.show()
```

```python
def ROCPlotter(pred, y_true):
    #Input: prediction probabilities and true labels
    #Output: ROC AUC curve
    
    lr_probs = pred
    ns_probs = [0 for _ in range(len(y_true))]
    testy = y_true
    ns_auc = roc_auc_score(testy, ns_probs)
    lr_auc = roc_auc_score(testy, lr_probs)
    print('Logistic: ROC AUC=%.3f' % (lr_auc))
    # calculate roc curves
    ns_fpr, ns_tpr, _ = roc_curve(testy, ns_probs)
    lr_fpr, lr_tpr, _ = roc_curve(testy, lr_probs)
    # plot the roc curve for the model
    plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
    plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')
    # axis labels
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    # show the legend
    plt.legend()
    # show the plot
    plt.show()
```
